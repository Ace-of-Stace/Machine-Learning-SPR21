# **Lecture: MNIST from Images (aka Multi-Classifier)**

In this notebook we will explore a classic data set, the MNIST data. This data set is a collection of hand written numerical characters 0 through 9. Each of the images is a single classified character, and each image is 28 by 28 pixels. Below we will read in the images, connect the feature and target data, and train a network to classify each of the characters. 

**New to this lecture:** A major new element that we will find in this notebook is a set of tools needed to classify multiple objects. Since we have 10 classes, we will need to develop a plan for how to train for all the different outputs. There are two major new tools that we need to watch for,

- to_categorical(Y_DATA)
- loss='categorical_crossentropy'

`to_categorical(Y_DATA)` allows us to One-Hot encode our targets, and `loss='categorical_crossentropy'` allows us to train for more than 2 targets. Below we will have notes and thoughts on these two new tools.
New "categories" are more multi-classifiers as opposed to binary classifiers like we've seen up until this point.

Another data tool that can be found in this lecture is a way to interact with images. In the imports, we will find `PIL` (Pillow). This allows us to edit pictures in the code (change colors, etc.)
This module will allow us to load images and turn them into arrays that we can use as data in a neural network. It should be noted that `Keras` does have some images tools as well, but `PIL` is nice to have so that we can develop a method of working with images outside of `Tensorflow`. 


````python
# Import tensorflow, import at the top to lok for issues.
import tensorflow as tf

# PIL (Pillow) for working with images; look up "Pillow" when checking for documantation.
import PIL
from PIL import Image , ImageOps

# Used for globbing up directories, glue together paths of directories and then search cummulatively.
import glob

# The Data Tools 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Import the Keras tools.
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras.utils import to_categorical

print("Tensorflow version:", tf.__version__) #Make sure to check versions, all revolves around GPUs Look at nvidia drivers and kudas on your computer
print("Pandas version:", pd.__version__)
print("Keras version:", keras.__version__)
print("Pillow version:",PIL.__version__,"\n")


physical_devices = tf.config.list_physical_devices('GPU') #Include this on all future notebooks.
print("GPU's:",physical_devices)
for gpu_instance in physical_devices: 
    tf.config.experimental.set_memory_growth(gpu_instance, True)
````

---
## **The Data**

````python
# The following block of code is going to read in each of the numbers from their directory, make
# them into an array, flatten them and normilize the values to be between 0 and 1. The loops 
# here will also append the target for each of the values.
# Dealing with images is challenging, prepare yourself. How to curate data sets for image analysis. 
# ** Set up image sets by putting target in the folders of that directory; target should be the directory it's in. 
# A range of the target integer values.
# You can find these images preprocessed in a csv sometimes.

NUM = np.arange(10) #Handwriten characters 0-9


# Empty lists for the features and targets. 
X_data  = []
Y_data = []

# Loop over all the target values (0 - 10)
for i in NUM:   
    
    # Glob up the file paths so that we can loop through all the images.
    ALL_IMAGES_TRAINING = glob.glob("../Data/education/MNIST/trainingSet/trainingSet/"+ str(i) +"/*.jpg") #Give it a directory and glue together to find images within a subdirectory
    #glob.glob grabs the path of the images
    # Loop over all the file paths. 
    for X in ALL_IMAGES_TRAINING: #X is a string that tells me where the images are
        
        # This line of code will open the image, make it an array, flatten the array, and divide all the values by 255.
        image = np.array(Image.open(X)).flatten() / 255 #Turn the image into an array and flatten (glue row to row to row to create a flat version
        # flatten command is important because it allows us to push an image into the network; every pixel is a feature after we 
        # Collect the features and the targets into the lists. #255 is the value for the scale grey scale or RGB
        X_data.append(image)
        Y_data.append(i)
````

````python
# Connect the features and the targets and then permute them.
DATA = np.random.permutation(np.concatenate((np.array(X_data), np.array([Y_data]).T), axis=1))
print(np.shape(DATA)) # Check the shape. 42000 columns, 785 rows
````

Next here we will take a look at on of the images. We will have to reshape it back to an 2 dim array since we flattened it above. 

````python
plt.matshow(DATA[3,:-1].reshape(28,28))
plt.show()
````

````python
# Break apart the features and targets. One "hot encoded vector" accepts tagets as integer values and returns one hot enoded vector
X_DATA = DATA[:,:-1]
Y_DATA = DATA[:,-1]
#set up a vector that has bins (locations) for each one; standard basis vectors (matrix all zeros except for one, taken from computer architecture of being on or off, 0 or 1)

#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!#
# Make the one-hot vectors for the categorical data. Note: from tensorflow.keras.utils import to_categorical is a Keras tool that
# allow us to turn the integer target values into one-hot vectors. 
train_labels = to_categorical(Y_DATA) #THIS IS THE IMPORTANT COMMAND!!
````

````python
print("This is a target:", Y_DATA[3])
print("This is a one-hot encoded target:", train_labels[3]) #This is the standard for multiclassifiers
````

In the next cell here we will break the data up into training, validation and testing data. Here the amounts are hardcoded and the validation data has been broken out from the data, but it should be noted that `Keras` can do this in the fit command. (Noted below)

````python
X_train = X_DATA[: int(np.shape(DATA)[0] * .8)] #Break into training, testing and validation data
Y_train = train_labels[: int(np.shape(DATA)[0] * .8)]

X_val = X_DATA[int(np.shape(DATA)[0] * .8) : int(np.shape(DATA)[0] * .9)] #Use this to prevent overtraining
Y_val = train_labels[int(np.shape(DATA)[0] * .8) : int(np.shape(DATA)[0] * .9)]

X_test = X_DATA[int(np.shape(DATA)[0] * .9) : ]
Y_test = train_labels[int(np.shape(DATA)[0] * .9) : ]
````

````python
# This cell will show a random data value and the one-hot target vector for it. 
n = np.random.randint(10000)
plt.matshow(X_train[n].reshape(28,28))
plt.show()
print("One hot target value:", Y_train[n])
````

---
## **The Network**

````python
# Setup the network layers. We are going to start to use RELU in place of sigmoid. 
# Here is link to the research as to why this change was made: http://proceedings.mlr.press/v15/glorot11a/glorot11a.pdf

# We will also switch to using softmax as the final activation since it will give us a probability distribution for
# our predicted output classes. This means that our output vectors values will sum to 1, and we can read each entry 
# as a probability. Softmax creates a probabilty distribution, how confident we are and next most confident and next most confident.

model = models.Sequential()
model.add(layers.Dense(200, activation='relu', input_shape=(np.shape(X_train[0])[0],), use_bias=False))
model.add(layers.Dropout(.5))
model.add(layers.Dense(100, activation='relu', use_bias=False))
model.add(layers.Dropout(.5)) #Way to code in a forgetfulness layer (forgets 50% of weights are rerandomized and gives network a mini-break, seems to speed up convergence, no trainable parameters)
#dropout is the day dreaming element.
model.add(layers.Dense(10, activation='softmax', use_bias=False))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) #Categorical cross entropy is the loss function to use with multi-classifiers
````

````python
model.summary()
````

---
## **The Training**

In the next cell we will train the network. We broke the data down so that we had a validation set, but the line of code here in this markdown cell shows a different option for building a validation set.
```Python
model.fit(X_train, Y_train, validation_split = .2, epochs=75, batch_size=100, verbose = 0)
```

````python
History = model.fit(X_train, Y_train, validation_data=(X_val, Y_val), epochs=400, batch_size=1000, verbose = 1) #verbose loads everything keras output and it's a lot.
# batch size determines how often we update the weights. Experiment with this. Epochs is number of times through the data (75-100 is good for this data set)
# Validation data broken off from main data, or can use a percent of the end (Keras does this)
# Below are the training and loss, how many right in training and validations data, HOWEVER, the validation data DOES NOT UPDATE the weights.
# This allows us to check that it works on data that the network was not necessarily train on. As the network sees the training data more and more, it should converge on accuracy
# But look for validation loss, as that grows, it means we are further from extrapolating the results to other data sets. = OVERTRAINING
````

## **Training Evaluation**

````python
plt.plot(History.history['loss']) #This allows us to check the number of epochs needed to train, but not overtrain. Look for elbow
plt.plot(History.history['accuracy'])
plt.xlabel("epoch")
plt.title("Training Data")
plt.show()

plt.plot(History.history['val_loss'])
plt.plot(History.history['val_accuracy'])
plt.xlabel("epoch")
plt.title("Validation Data") # We do not want the validation loss to increase, it will start to creep up. When we get to the point when we see increase here, that's the cut off for training
plt.show()
````

## **The Testing Data**

````python
Points = np.array(list(zip(np.argmax(model.predict(X_test), axis=1), Y_DATA[int(np.shape(DATA)[0] * .9):])))
````

````python
Unique_Points, Unique_Counts = np.unique(Points, axis=0, return_counts=True)
plt.scatter(Unique_Points[:,0], Unique_Points[:,1], 20 / (Unique_Counts) , c = -20 / Unique_Counts) #Suppressed correct guesses
plt.xlabel("Guess")
plt.xticks(NUM)
plt.ylabel("Target")
plt.yticks(NUM)
plt.title("The Mistakes")
plt.show() #Graphic shows biggest offenders is missed guesses
````

````python
# Check to see how many were correct in the testing data.
Correct = np.unique(np.argmax(model.predict(X_test),axis=1) - Y_DATA[int(np.shape(DATA)[0] * .9):], return_counts= True)[1][9]
````

````python
# Percent correct.
(Correct / np.shape(Y_DATA[int(np.shape(DATA)[0] * .9):])[0]) * 100 #This bad percentage below is a result of over-training
````

## **Production**
In the next few blocks of code we will read in a new character that I wrote and have the network determine what the character is. We will need to remove any extra color layers that it may have and normalize the entries. These data "reworks" will need to be built into any production tool that we deploy this network with.

````python
# Read in the image and grayscale it.
image = Image.open("3.jpg")
image = ImageOps.grayscale(image)

# Flip the colors and normilize the values.
im_array = (255 - np.array(image)) / 255

# Display the character to see that it looks right.
plt.matshow(im_array)
plt.show()
````

````python
# Flatten it to feed it to our network (nom nom)
flat = np.array([im_array.flatten()])
````

````python
# Check what value that has the highest probability and what that probability is.
print("The network thinks the value is," , np.argmax(model.predict(flat)), "with a probability of," , np.max(np.round(model.predict(flat),1).T ))
````

````python
# Plot of the probabilities for each of the targets.
plt.scatter(NUM,np.round(model.predict(flat),1).T )
plt.xlabel("Target Numbers")
plt.ylabel("Probabilities")
plt.show()
````
