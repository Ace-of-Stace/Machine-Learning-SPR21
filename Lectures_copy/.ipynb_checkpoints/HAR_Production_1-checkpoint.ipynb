```python
import pandas as pd #Alias for pandas is pd
print(pd.__version__)
import numpy as np
import matplotlib.pyplot as plt

print(plt.rcParams.get('figure.figsize'))
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size
```

```python
S = pd.read_csv("Collab/stacie/AH_Data/AH_Static_Sensor_Data/AH_Walking_Sensor.csv") # Read in the walking data. 
S # Show the dataframe
```

```python
print(S.keys()) #Goes to A dataframe and finds the keys
```

```python
time = S["time"]

x_acceleration = S["xAcc"]
y_acceleration = S["yAcc"]
z_acceleration = S["zAcc"]

plt.plot([],[], color='r', label = 'x_acceleration')
plt.plot([],[], color='g', label = 'y_acceleration')
plt.plot([],[], color='b', label = 'z_acceleration')

plt.stackplot(time, x_acceleration, y_acceleration, z_acceleration, colors = ['r','g','b'])
plt.legend()
plt.title('Walking Sensor data')
plt.xlabel('time')
plt.ylabel('Acceleration')
plt.show()
```

```python
time = S["time"]

x_gyro = S["xGyro"]
y_gyro = S["yGyro"]
z_gyro = S["zGyro"]

plt.plot([],[], color='r', label = 'x_gyro')
plt.plot([],[], color='g', label = 'y_gyro')
plt.plot([],[], color='b', label = 'z_gyro')

plt.stackplot(time, x_gyro, y_gyro, z_gyro, colors = ['r','g','b'])
plt.legend()
plt.title('Walking Gyro data')
plt.xlabel('time')
plt.ylabel('Gyro')
plt.show()
```

```python
time = T["time"]

x_acceleration = T["xAcc"]
y_acceleration = T["yAcc"]
z_acceleration = T["zAcc"]

plt.plot([],[], color='r', label = 'x_acceleration')
plt.plot([],[], color='g', label = 'y_acceleration')
plt.plot([],[], color='b', label = 'z_acceleration')

plt.stackplot(time, x_acceleration, y_acceleration, z_acceleration, colors = ['r','g','b'])
plt.legend()
plt.title('Sit To Lie data')
plt.xlabel('time')
plt.ylabel('Acceleration')
plt.show()
```

```python
time = T["time"]

x_gyro = T["xGyro"]
y_gyro = T["yGyro"]
z_gyro = T["zGyro"]

plt.plot([],[], color='r', label = 'x_gyro')
plt.plot([],[], color='g', label = 'y_gyro')
plt.plot([],[], color='b', label = 'z_gyro')

plt.stackplot(time, x_gyro, y_gyro, z_gyro, colors = ['r','g','b'])
plt.legend()
plt.title('Sit To Lie Gyro data')
plt.xlabel('time')
plt.ylabel('Gyro')
plt.show()
```

```python
fig = plt.figure(figsize=(20,20))
ax = plt.axes(projection='3d')

plt.title("X Gyro vs Y Gyro vs Z Gyro")
plt.xlabel("X Gyro")
plt.ylabel("Y Gyro")
ax.set_zlabel('Z Gyro')


X = T["xGyro"]
Y = T["yGyro"]
z = T["zGyro"]

ax.scatter3D(X, Y ,z , c = z, marker='o')
ax.view_init(elev = 0, azim=120)
plt.show()
```

```python
contour_data = pd.read_csv("Collab/stacie/AH_Data/AH_Transition_Sensor_Data/AH_Sit_To_Lie_Sensor.csv")
contour_data.head()

Z = contour_data.pivot_table(index='xGyro', columns='yGyro', values='zGyro').T.values

X_unique = np.sort(contour_data.xGyro.unique())
Y_unique = np.sort(contour_data.yGyro.unique())
X, Y = np.meshgrid(X_unique, Y_unique)
```
